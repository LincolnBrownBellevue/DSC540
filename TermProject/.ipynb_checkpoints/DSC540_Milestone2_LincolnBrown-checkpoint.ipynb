```python
# Import libraries, set display rows
import pandas as pd
import numpy as np
pd.options.display.max_rows = 999
```

```python
# Read in the dataset
chicago_crime_df = pd.read_csv("ChicagoCrime.csv")
print(chicago_crime_df.columns)
chicago_crime_df.head(10)
```

```python
# Number of rows and columns in flat file dataset
df_shape = chicago_crime_df.shape

print(f"Number of rows: {df_shape[0]}")
print(F"Number of columns {df_shape[1]}")
```


## Step 1

The column names contain unnecessary leading spaces and I also want to remove all spaces and replace them with underscores. There is also a column with 2 spaces where there should only be one

## Step 2

I also don't like all capital letters for column names, so I'll put them in title case.


```python
cols = chicago_crime_df.columns
clean_cols = []
for col in cols:
    col = col.lstrip()
    col = col.replace("  ", "_")
    col = col.replace(" ", "_")
    if(col == 'IUCR'):
        pass
    else:
        col = col.title()
    clean_cols.append(col)

print(clean_cols)
chicago_crime_df.columns = clean_cols
chicago_crime_df
```


## Step 3

The date column is actually a date time column. I want to separate the date and time into two separate columns

```python
chicago_crime_df['Date_Of_Occurrence'] = pd.to_datetime(chicago_crime_df['Date_Of_Occurrence'])
chicago_crime_df['Date'] = chicago_crime_df['Date_Of_Occurrence'].dt.date
chicago_crime_df['Time'] = chicago_crime_df['Date_Of_Occurrence'].dt.time
chicago_crime_df
```

## Step 4 
The X and Y coordinate columns refer to the location where the incident occurred in State Plane Illinois East NAD 1983 projection. This location is shifted from the actual location for partial redaction but falls on the same block.
Since I don't know where the State Plane Illinois East NAD 1983 projection is, I am going to drop these columns. 

```python
chicago_crime_df.drop(columns=['X_Coordinate', 'Y_Coordinate'], inplace=True)
chicago_crime_df
```

## Step 6

I am going to clean up the dataset because the rows that have NaN can't simply be dropped.

```python
# From the snippet above, we can see that any Secondary Description that spanned two rows
# is recognized as a new row in the dataset
# Correcting this will be my first step. 
# Realistically, I don't need to do this, because I can just match the IUCR code to my
# chicago_crime_df dataset and then use those descriptions, but I want the practice 

# How many lines are NA in iucr_codes
nas = iucr_codes.isna().sum()
rows = iucr_codes.shape[0]
cleaned_rows = rows - nas[0]
print(f"Cleaned dataset should have {cleaned_rows} rows")
#print(f"Final dataset should have {total_clean} rows.\n\n")
test = iucr_codes.iloc[42, 0]
#print(test)

test2 = iucr_codes.iloc[369]
#print(test2)

#print(test2['PRIMARY DESCRIPTION'])
#print(type(test))

# There is probably a way to do this with a one-liner (maybe two)
# But this is how I'm going to approach the problem
# First, we'll iterate through the rows in the dataframe, looking for NaNs
# Then we'll parse which column contains the data
# and append it to the row's value before it
for index, row in iucr_codes.iterrows():
    #print(row['IUCR'])
    if(type(row['IUCR']) == float and pd.isna(row['IUCR'])):
        #print(row['SECONDARY DESCRIPTION'])
        if(type(row['SECONDARY DESCRIPTION'])) == str:
            idx = index - 1
            desc = iucr_codes.iloc[idx, 2]
            full_desc = f"{desc} {row['SECONDARY DESCRIPTION']}"
            iucr_codes.iloc[idx, 2] = full_desc
        elif(type(row['PRIMARY DESCRIPTION'])) == str:
            idx = index - 1
            desc = iucr_codes.iloc[idx, 1]
            full_desc = f"{desc} {row['PRIMARY DESCRIPTION']}"
            iucr_codes.iloc[idx, 1] = full_desc

# Going to make a new df to keep it separate
clean_iucr = iucr_codes.dropna()
print(f"clean_iucr has {clean_iucr.shape[0]} rows.")
clean_iucr
```

## Step 7

Now I'm going to merge the two dataframes together on the IUCR so that 
chicago_crimes_df will have a Index Code column
Then I will use that to create a new DataFrame of index crimes that can be used later

```python
# Now that the dataset is clean, let's use the 'INDEX CODE' column to
# add a new column to the chicago_crime_df 
index_crimes = pd.merge(chicago_crime_df, clean_iucr[['IUCR','INDEX CODE']], how='inner', on='IUCR') 
tracked_crimes = index_crimes.loc[index_crimes['INDEX CODE'] == 'I']
tracked_crimes
```

## Ethical Considerations

Any time you are cleaning data it is important to keep in mind the ethical considerations. It is especially important when dealing with crime. Handling data pertaining to crime is an easy place to misrepresent statistics and information that can have far-reaching effects on a city's economy, populace, and safety. It is important to keep this in mind so that all data are cleaned reliably, any steps taken to clean the data are recorded, and that all findings are reproducible.
