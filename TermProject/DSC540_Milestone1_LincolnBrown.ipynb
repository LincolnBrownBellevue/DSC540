## API Dataset

The final dataset is the Weather API. I looked at numerous sites for this data and found the website Visual Crossing will meet my needs:
https://www.visualcrossing.com/weather/weather-data-services#

They allow up to 1000 records per day and after some modifications, I was able to download JSON records for each month I require.

```python
import json

# I will just load one month in so the dataset can be observed
json_file = "TermProject/WeatherData/09_2022.json"
with open(json_file, "r") as file:
    data = file.readlines()

json_data = ""
# Send it over to json.loads to make it easier to work with
for item in data:
    json_data = json.loads(item)

# First dataframe
json_df = pd.DataFrame(json_data)
print(f"Number of rows: {json_df.shape[0]}")
print(f"Number of columns: {json_df.shape[1]}")
print(json_df.columns)
# Well those columns don't look like the data I'm after.
# It is buried, so I need to extract the JSON I'm looking for.

# Each day's weather data resides as a series in the 'days' column
json_df['days']
sept_days = []
# Create a list so that I don't need to concatenate a bunch of df's together
for item in json_df['days']:
    sept_days.append(item)

# Using a list of dicts, I will create a dataframe for September 2022
Sept22_df = pd.DataFrame.from_dict(sept_days, orient='columns')
print(Sept22_df.head())
print(f"Number of rows: {Sept22_df.shape[0]}")
print(f"Number of columns: {Sept22_df.shape[1]}")
```
