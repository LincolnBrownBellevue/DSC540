## Load the dataframes into SQLite database


```python
# Create a function to check the status of the database connection
def check_connection(conn):
    try:
        conn.cursor()
        print("Connection Sucessful")
    except Exception as e:
        print("Connection Failed")
   
```

```python
# Connect to the database and verify it is connected 
database_file = "DSC540_Project.db"
with sqlite3.connect(database_file) as conn:
    cursor = conn.cursor()
    check_connection(conn)
```

```python
def dataframe_reindexing(dataframe):
    # Create id column from index
    dataframe = pd.concat([pd.Series(dataframe.index, index=dataframe.index, name='id'), dataframe], axis=1) 
    return dataframe
```

```python
# Create build_table SQL query string
def build_table(dataframe, table_name):
    # Build query specifically in this function
    dataframe_cols = dataframe.columns
    print(dataframe_cols)
    build_table_query = f"CREATE TABLE IF NOT EXISTS {table_name} (\n"
    for index, col in enumerate(dataframe_cols):
        last_item = len(dataframe_cols) -1
        if col.lower() == 'id':
            build_table_query += f"\tid integer PRIMARY KEY,\n"
        elif index < last_item:
            build_table_query += f"\t{col.lower()} text,\n"
        else:
            build_table_query += f"\t{col.lower()} text\n);"
    return build_table_query
```

## Chicago Crime Dataset

```python
# Drop table if it exists
cursor.execute("DROP TABLE IF EXISTS chicago_crime")
conn.commit()
```

```python
# Confirm there are no tables
tables = cursor.execute("SELECT * FROM sqlite_master WHERE TYPE='table'")
for table in tables:
    print(table)
```

```python
# Rename unnamed column to id to use for primary key
del chicago_crime_df['Unnamed: 0']
chicago_crime_df['id'] = chicago_crime_df.index
chicago_crime_df.rename(columns = {'Case#': 'case_num'}, inplace=True)
# Create a dictionary to hold the value places in the upcoming sql query
crime_cols = chicago_crime_df.columns
crime_dict = {}
for col in crime_cols:
    crime_dict[col.lower()] = '?'
```

```python
# Create the sql query to build the chicago_crime table
build_crime_table = "CREATE TABLE IF NOT EXISTS chicago_crime (\n"
for index, col in enumerate(crime_cols):
    last_item = len(crime_cols) -1
    if col == 'id':
        build_crime_table += f"\tid integer PRIMARY KEY,\n"
    elif index < last_item:
        build_crime_table += f"\t{col.lower()} text,\n"
    else:
        build_crime_table += f"\t{col.lower()} text\n);"

print(build_crime_table)
```

```python
# Create a dictionary of the columns and str type for conversion
# Create new dataframe that will be used to create the SQL table
convert_dict = {}
for col in chicago_crime_df.columns:
    convert_dict[col] = str
sql_crime_df = chicago_crime_df.astype(convert_dict)
sql_crime_df.dtypes
```

```python
# Confirm the dataframe looks correct
sql_crime_df
```

```python
# Create the table and commit the changes
cursor.execute(build_crime_table)
conn.commit()
```

```python
# Insert the sql_crime_df dataframe onto the table
insert_crime_query = f"INSERT OR REPLACE INTO chicago_crime ({', '.join(map(str,crime_dict.keys()))}) values ({', '.join(map(str,crime_dict.values()))})"  
print(insert_crime_query)
cursor.executemany(insert_crime_query, sql_crime_df.to_records(index=False))
conn.commit()
```

```python
# Try to read the table to make sure it was successful
crime_table = cursor.execute("SELECT * FROM chicago_crime")

for index, row in enumerate(crime_table):
    if index < 10:
        print(row)
```
