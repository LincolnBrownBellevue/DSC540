### Clean the flat file
Perform at least 5 data transformation and/or cleansing steps to your flat file data

```python
# Import libraries, set display rows
import pandas as pd
import numpy as np
pd.options.display.max_rows = 999
```

```python
# Read in the dataset
chicago_crime_df = pd.read_csv("./TermProject/ChicagoCrime.csv")
print(chicago_crime_df.columns)
chicago_crime_df.head(10)
```

```python
# Number of rows and columns in flat file dataset
df_shape = chicago_crime_df.shape

print(f"Number of rows: {df_shape[0]}")
print(F"Number of columns {df_shape[1]}")
```


## Step 1

The column names contain unnecessary leading spaces and I also want to remove all spaces and replace them with underscores. There is also a column with 2 spaces where there should only be one

## Step 2

I also don't like all capital letters for column names, so I'll put them in title case.


```python
cols = chicago_crime_df.columns
clean_cols = []
for col in cols:
    col = col.lstrip()
    col = col.replace("  ", "_")
    col = col.replace(" ", "_")
    if(col == 'IUCR'):
        pass
    else:
        col = col.title()
    clean_cols.append(col)

print(clean_cols)
chicago_crime_df.columns = clean_cols
chicago_crime_df
```


## Step 3

The date column is actually a date time column. I want to separate the date and time into two separate columns

```python
chicago_crime_df['Date_Of_Occurrence'] = pd.to_datetime(chicago_crime_df['Date_Of_Occurrence'])
chicago_crime_df['Date'] = chicago_crime_df['Date_Of_Occurrence'].dt.date
chicago_crime_df['Time'] = chicago_crime_df['Date_Of_Occurrence'].dt.time
chicago_crime_df
```

## Step 4 
The X and Y coordinate columns refer to the location where the incident occurred in State Plane Illinois East NAD 1983 projection. This location is shifted from the actual location for partial redaction but falls on the same block.
Since I don't know where the State Plane Illinois East NAD 1983 projection is, I am going to drop these columns. 

```python
chicago_crime_df.drop(columns=['X_Coordinate', 'Y_Coordinate'], inplace=True)
chicago_crime_df
```

```python

print(chicago_crime_df.columns)

print(chicago_crime_df.isna().any())
```


## Step 5

I am going to drop the rows containing NA's in the Location


```python
chicago_crime_df.dropna(subset=['Location'], inplace=True)
print(chicago_crime_df.shape)
chicago_crime_df.isna().any()
```
