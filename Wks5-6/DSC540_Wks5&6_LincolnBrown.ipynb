```python
# Step 1
# Open the Wikipedia Page
from bs4 import BeautifulSoup as bs
import pandas as pd
import re
import requests

url = "https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)"
r = requests.get(url)
r.text
```

```python
# Step 2: Read the page using BeautifulSoup
# Convert the request page into soup
soup = bs(r.text, 'lxml')
# Much easier to read
print(soup)
```

```python
# Step 3: Find the table structure you will need to deal with (how many tables are there?)
tables = soup.find_all("table")
print(f"There are {len(tables)} tables in this page.")
```

```python
# Step 4 Find the right table using bs4
# We're looking for the class 'wikitable'
for table in tables:
    print(table.attrs)
    #print(table.attrs.keys())
    #print(table.attrs.values())
    
#print(wikitable)
wikitable = soup.find("table", {"class": "wikitable"})
print(wikitable)
```

```python
# Step 5 Separate the source names and their corresponding data
# Basically, isolate IMF, World Bank, and UN source lists so that we can
# scrape the data from each table
# Grab the first row of the table and put it into sources var
sources = wikitable.tbody.findAll('tr')[0]
# Webpage has these set as headers instead of table data now so need to search 'th'
sources_list = [td for td in sources.findAll('th')]
print(len(sources_list))
#print(sources_list)
# Since the webpage has been updated we are pulling 5 columns, but we only need the last 3
# Need to parse out those that we don't want
# All of the columns we want have a number, so we'll use that
# Store the cleaned names for later and remove unwanted sources from the list
clean_sources = {}
for source in sources_list:
    if(re.search(r'\d', source.text)):
        pattern = re.compile(r'[\W\d]')
        clean_str = re.sub(pattern, '', source.text)
        print(clean_str)
        clean_sources[clean_str] = source
assert(len(clean_sources) == 3)
```

```python
#Clean GDP
def clean_gdp(index, td):
    if(index == 0):
        # Get Rank
        return td.text.strip()
    elif(index == 1):
        # Get Country name
        return td.text.strip()
    else:
        # Clean GDP str
        i = td.text.find("♠")
        return td.text[i+1:].strip()
```

```python
# Showing the cleaned output
wb_cleaned_data = [[clean_gdp(index, td) for index, td in enumerate(tr.findAll('td'))] for tr in wb_rows[2:]]
print(wb_cleaned_data)
```

```python
# Add it to a dataframe
wb_df = pd.DataFrame(wb_cleaned_data, columns=wb_headers)
wb_df
```
